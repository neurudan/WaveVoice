```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
path = '/media/bratwolf/DATA/results_cluster/'
extension = '/logs.csv'

t_accs = []
v_accs = []
t_losses = []
v_losses = []

[fig, [[ax0, ax1], [ax2, ax3]]] = plt.subplots(2,2, figsize=(18, 12))
labels = []

for dir in os.listdir(path):
    file = path + dir + extension
    
    df = pd.read_csv(file)

    epochs = df['epoch'].to_numpy()

    t_acc = df['accuracy'].to_numpy()
    t_loss = df['loss'].to_numpy()

    v_acc = df['val_accuracy'].to_numpy()
    v_loss = df['val_loss'].to_numpy()

    ax0.plot(epochs, t_acc)
    ax1.plot(epochs, v_acc)
    ax2.plot(epochs, t_loss)
    ax3.plot(epochs, v_loss)

    labels.append(dir)
ax0.set_title('train accuracy')
ax1.set_title('val accuracy')
ax2.set_title('train loss')
ax3.set_title('val loss')
fig.legend(tuple(labels))
```
