```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot(file):
    df = pd.read_csv(file)

    epochs = df['epoch'].to_numpy()

    t_acc = df['accuracy'].to_numpy()
    t_loss = df['loss'].to_numpy()

    v_acc = df['val_accuracy'].to_numpy()
    v_loss = df['val_loss'].to_numpy()

    [_, [ax0, ax1]] = plt.subplots(1,2, figsize=(18, 6))
    print(ax0)
    ax0.plot(epochs, t_acc, epochs, v_acc)
    ax1.plot(epochs, t_loss, epochs, v_loss)
```

```python
plot('/media/bratwolf/DATA/results_cluster/byte_net/logs.csv')
```
